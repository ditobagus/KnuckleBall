```python
import numpy as np
from matplotlib import pyplot as plot
import math
from mpl_toolkits import mplot3d
```

```python
res = 0
#drag coefficient of sphere object
drag_coefficient = .47
lift_coefficient = 0
gravity = 9.8
velocity_of_ball = 0
#speed of air above the ball
upper_velocity = 10
#speed of air below the ball
lower_velocity = 20
#air density
p = .0012
wind_velocity = 15
vector_of_velocity = np.array([[10, 5, 15],
                      [12, 7, 17],
                      [14, 9, 19],
                      [12, 7, 15],
                      [10, 5, 15]
                     ])
vector_of_force = np.array([[127, 51, 301],
                      [193, 73, 359],
                      [243, 101, 421],
                      [181, 92, 376],
                      [152, 63, 331]
                     ])
```

```python
#Soccer ball

velocity_of_ball = 51
M = 210
#diameter
D = 21
#Surface area
S = 4 * math.pi * ((D/2) ** 2)
#Lift force
L = .5 * p * (lower_velocity ** 2 - upper_velocity ** 2) * S
lift_coefficient = (2 * L) / (p * (wind_velocity ** 2) * S)
#numerical factor from mass, radius and air density
K = ((4/3) * math.pi * ((D/2) ** 3) * p) / M
res = K * (-1) * drag_coefficient * (velocity_of_ball ** 2) * vector_of_velocity + K * lift_coefficient * (velocity_of_ball ** 2) * (vector_of_velocity * vector_of_force) + gravity
```

```python
#plot disini
fig = plot.figure()
axis = plot.axes(projection='3d')
```

```python
fig = plot.figure()
axis = fig.add_subplot(111, projection='3d')
axis.plot(res[0],res[1],res[2])
plot.show()
```
